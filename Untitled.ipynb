```python
# fraud_detection_no_shap.ipynb

# Import libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import lightgbm as lgb
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, classification_report, roc_curve, precision_recall_curve
```

```python
# Load your dataset (adjust path as needed)
# Folder where your .pkl files are stored
folder_path = './dataset/data'
# List all .pkl files
pkl_files = [file for file in os.listdir(folder_path) if file.endswith('.pkl')]

# Load all files and concatenate into one DataFrame
all_dfs = []

for file in sorted(pkl_files):  # sort to maintain date order
    file_path = os.path.join(folder_path, file)
    df = pd.read_pickle(file_path)
    all_dfs.append(df)

# Combine all into one DataFrame
fraud_data = pd.concat(all_dfs, ignore_index=True)

# Preview combined data
print(fraud_data.shape)
print(fraud_data.head())
```

```python
fraud_data['CUSTOMER_ID'] = fraud_data['CUSTOMER_ID'].astype('int')
```

```python
# import pandas as pd

# Ensure datetime format
fraud_data['TX_DATETIME'] = pd.to_datetime(fraud_data['TX_DATETIME'])

# Sort first
fraud_data = fraud_data.sort_values(['CUSTOMER_ID', 'TX_DATETIME'])

# -----------------------------
# CUSTOMER FEATURE: rolling tx count in 7 days
# -----------------------------
def customer_tx_count(group):
    group = group.set_index('TX_DATETIME')
    group['customer_tx_count_7d'] = group['TRANSACTION_ID'].rolling('7D').count()
    group = group.reset_index()  # Bring TX_DATETIME back as a column
    return group

fraud_data = fraud_data.groupby('CUSTOMER_ID').apply(customer_tx_count,include_groups=False).reset_index(drop=True)

# ðŸŸ¢ Now TX_DATETIME is safely a column again

# -----------------------------
# TERMINAL FEATURE: rolling fraud ratio in 7 days
# -----------------------------
fraud_data = fraud_data.sort_values(['TERMINAL_ID', 'TX_DATETIME'])

def terminal_fraud_ratio(group):
    group = group.set_index('TX_DATETIME')
    total_tx = group['TX_FRAUD'].rolling('7D').count()
    fraud_tx = group['TX_FRAUD'].rolling('7D').sum()
    group['terminal_fraud_ratio_7d'] = fraud_tx / total_tx
    group = group.reset_index()  # Again bring TX_DATETIME back
    return group

fraud_data = fraud_data.groupby('TERMINAL_ID').apply(terminal_fraud_ratio,include_groups=False).reset_index(drop=True)
```

```python

# Prepare data for modeling
features = ['TX_AMOUNT', 'customer_tx_count_7d', 'terminal_fraud_ratio_7d', 'tx_hour']
X = fraud_data[features]
y = fraud_data['TX_FRAUD']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Train LightGBM model
lgb_train = lgb.Dataset(X_train, label=y_train)
lgb_eval = lgb.Dataset(X_test, label=y_test, reference=lgb_train)

params = {
    'objective': 'binary',
    'metric': 'auc',
    'boosting_type': 'gbdt',
    'scale_pos_weight': (len(y_train) - sum(y_train)) / sum(y_train),  # balance class weight
    'verbose': -1
}

model = lgb.train(
    params,
    lgb_train,
    valid_sets=[lgb_train, lgb_eval],
    callbacks=[
        lgb.early_stopping(50),
        lgb.log_evaluation(10)  # replaces verbose_eval=10
    ]
)

# Predict and evaluate
y_pred_proba = model.predict(X_test, num_iteration=model.best_iteration)
y_pred = (y_pred_proba >= 0.5).astype(int)

print("ROC AUC Score:", roc_auc_score(y_test, y_pred_proba))
print(classification_report(y_test, y_pred))

# Plot ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, label='LightGBM (area = %0.3f)' % roc_auc_score(y_test, y_pred_proba))
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='best')
plt.savefig("ROC Curve.png", dpi=300, bbox_inches='tight')
plt.show()

# Plot feature importance
lgb.plot_importance(model, max_num_features=10, importance_type='gain', figsize=(8,6))
plt.title('LightGBM Feature Importance')
plt.savefig("LightGBM Feature Importance.png", dpi=300, bbox_inches='tight')
plt.show()
```
